# Ataque zombie!

```python
%pylab inline
```

```python
import rungekutta

s0 = 750 # Humanos iniciales
z0 = 1
r0 = 0

alpha = 0.005 # Natalidad humanos
beta = 0.0006 # Infección
delta = 0.0001 # Muerte natural humanos
xi = 0.000 # Resurrección
alpha2 = 0.0005 # Muerte por humanos de los zombies
sigma = 0.004 # Muerte natural zombies
capacidad = 10000

tmax = 500

sol = rungekutta.runge_kutta(s0,z0,r0,tmax,[alpha,beta,delta,xi,alpha2,sigma,capacidad])
```

```python
plot(linspace(0,tmax*10+2,tmax*10+2),sol[0],color="blue",label="Humanos")
plot(linspace(0,tmax*10+2,tmax*10+2),sol[1],color="red",label="Zombies")
plot(linspace(0,tmax*10+2,tmax*10+2),sol[2],color="black",label="Muertos")
legend()
show()
```

```python
%timeit rungekutta.runge_kutta(s0,z0,r0,tmax,[alpha,beta,delta,xi,alpha2,sigma,capacidad])
```
